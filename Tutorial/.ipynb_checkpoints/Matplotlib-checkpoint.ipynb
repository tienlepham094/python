### Cách vẽ matplotlib
#### Pyplot API vs Object-Oriented (OO) API
- Pyplot APi -> quickly
- OO API -> advanced

```python
x = np.linspace(0,10,1000)

plt.plot(x, np.sin(x), color="blue", linestyle="dashed")
plt.plot(x, np.cos(x),color="red", label="cos(x)")


# đặt label cho đồ thị
plt.xlabel('Bien x')
plt.ylabel('Sin(x)')

# giới hạn đồ thị
# plt.xlim([0,4])

plt.legend() # hiện label

# đặt title cho đồ thị
plt.title("Sin(x) and Cos(x)")
```

## Object-Oriented API

```python
x = [1,2,3,4]
y = [11,22,33,44]

#set up plot
fig, ax = plt.subplots(figsize=(5,5)) #figure size = width & height of the plot -> customize size

# plot the data
ax.plot(x,y)

# set ttile
ax.set(title="A simple plot",
      xlabel="x-axis",
      ylabel="y-axis")
```

### 1. Line Chart

### 2. Scatter (Biểu đồ phân tán)

```python
# Python API
plt.scatter(x, np.exp(x)) # y= e^x
```

```python
## OOP Api

#Prepare new data
rng = np.random.RandomState(0)

x = rng.randn(100)
y =rng.randn(100)

fig, ax = plt.subplots()
ax.scatter(x,y)
```

### 3. Bar chart

- Veritcal
- Horizontal 

```python
# Prepare data from a python dictionary
soft_drink_prices={"Coke":10, "Pepsi": 12,
                  "Sprite": 8}

fix,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax.set(title="Soft Drink Prices")
```

```python
fix, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),
        list(soft_drink_prices.values()))
```

### 4. Histogram

```python
# prepare data
student_height = np.random.normal(170,10,250) # median: 170, dao động: -+10, số: 250
```

```python
student_height[::10]
```

```python
# pylot API
plt.hist(student_height)
```

### Subplots
Vẽ nhiều đố thị
