# Sales Analytics

#### Import Neccessary Libraries

```python
import pandas as pd
import os
```

###  Task #1: Merging 12 moths of sales data into a single file

```python
# listdir trong python trả về danh sách chứa tên các thư mục trong path đã cho
# os.listdir(path)

files = [file for file in os.listdir('./Folder_csv')  if not file.startswith('.')]
# print(files)

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Folder_csv/'+ file)
    
    # concat in pandas
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index = False)
```

#### Read in updated DataFrame

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

#### Clean data (drop rows of NAN)

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### Adding additional columns

### Task #2: Add Month Column

```python
# convert column
all_data['Order Date'] =pd.to_datetime(all_data['Order Date'],errors='coerce' )
all_data["Month"] = all_data['Order Date'].dt.month

all_data.head()
```

#### Convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Question 1: What was the best moth for sales? How much was earned that moth?

```python
all_data["Month"].dtype
```

```python
all_data.groupby('Month').sum()
```

### Task #4: Add a city column 

```python
# use apply method
all_data['City'] = all_data['Purchase Address'].apply(lambda x : x.split(',')[1])
all_data.head()
```

#### Question 2: What city had the highest number of sales ? 

```python
all_data.groupby('City').sum()
```

```python
all_data[['City', 'Sales']].groupby('City').sum()
```

#### Question 3: What prodcuts are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df.head()
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df
```

```python
grouped = df.pivot_table(columns=['Grouped'],aggfunc='size')
print(grouped)
```
