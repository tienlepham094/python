# Task #1: Get info Box (store in Python dictionary)

### Import Necessary Libraries

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

### Load the webpage

```python
r = requests.get("https://en.wikipedia.org/wiki/Toy_Story_3")

# Convert to a beautiful soup object
soup = bs(r.text)

# Print out the HTML
contents = soup.prettify()
# print(contents)
```

```python
info_box = soup.find("table", {"class": "infobox vevent"})
info_rows = info_box.findAll("tr")
```

```python
movie_info = {}

# function to transform data into array
def get_content_value(row_data):
    if row_data.find("li"):
        array=[]
        for li in row_data.find_all("li"):
            array.append(li.text)
        return array
    else:
        return row_data.text.strip()
        
# apeend to dictionary
for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['title'] = row.find("th").text
    elif index ==1:
        continue
    else:
        content_key = row.find("th").text.strip()
        content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value

movie_info
```

# Task #2: movies in List of Disney Films

```python
r2 = requests.get("https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films")

# Convert to beautifulSoup Object
soup_2 = bs(r2.text)
# print(soup_2.prettify())
```

```python
movie_table = soup_2.findAll("table", {"class": "wikitable sortable"})
# print(movie_table)
movies = movie_table[0:3]
# print(movies)
```

extract table

```python
def extract_table(table):
    dataframe = pd.DataFrame(columns=["Title", "Release Date", "Notes"])
    rows = table.findAll("tr")
    for index, row in enumerate(rows):
        if index == 0:
            continue
        else:
            col = row.find_all('td')
            title = col[1].text.strip()
            date = col[2].text.strip()
            notes = col[3].text.strip()
            dataframe = dataframe.append({"Title": title, "Release Date": date, "Notes": notes}, ignore_index = True)
    return dataframe
extract_table(movies[0])
            
            
```

```python
for movie in movies:
    df = pd.DataFrame(columns=["Title", "Release Date", "Notes"])
    df = df.append(extract_table(movie))
df
```
